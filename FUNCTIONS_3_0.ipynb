```python
#1. 1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in
#    the list.
def sum_even(a):
  n=0
  for i in a:
      if i%2==0:
        n+=i
      else:
        continue
  return n
sum_even([1,2,3,4,5,6,7])


```

```python
#2 ***  Create a Python function that accepts a string and returns the reverse of that string.
def rev_string(a):
 # len= len(a) - 1
  b=" "
  for i in range(len(a)-1,-1,-1):
    b = b + str(a[i])+","

  return b

rev_string((1,2,3,4,5,6,7))

```

```python
#3 Implement a Python function that takes a list of integers and returns a new list containing the squares of each number.

def list_squ(a):
  n=[]
  for i in a:
    n.append((i)**2)
  return n

list_squ([1,2,3,4,5])

```

```python
#
def is_prime(num):
  if num<2:
    return False
  for i in range(2,int(num**0.5)+1):
    if num%i==0:
      return False
  return True

for a in range(1, 200+1):
  if  is_prime(a):
    print(f"{a} is prime Number")
  else:
    print(f"{a} is Not a prime Number")



```

```python
#6  Write a generator function in Python that yields the powers of 2 up to a given exponent.
def power_of_two(n):
  for i in range(n+1):
    yield 2**i

n=10
'''print(list(power_of_two(n)))'''
for power in power_of_two(n):
  print(power, end = ",")


```
